```python
import seaborn as sns

from analytix import Analytics
```

```python
client = Analytics.with_secrets("secrets.json")
report = client.retrieve(
    dimensions=("day",),
    metrics=(
        "views",
        "estimatedMinutesWatched",
        "subscribersGained",
        "subscribersLost",
        "estimatedAdRevenue",
    ),
)
```

```python
df = report.to_dataframe()
print(df.head())
```

```python
sns.lineplot(x="day", y="views", data=df)
```

```python
sns.lineplot(x="day", y="estimatedMinutesWatched", data=df)
```

```python
sns.lineplot(x="day", y="subscribersGained", data=df)
sns.lineplot(x="day", y="subscribersLost", data=df)
```

```python
sns.lineplot(x="day", y="estimatedAdRevenue", data=df)
```
